## STEP 4: Correlation Analysis and Hypothesis testing  
- **Hypothesis testing** is a statistical method used to determine whether there's enough evidence in data to support a claim about a r/ship between variables.  
- Hypothesis testing helps us:
    - Feature Selection  
    - EDA(To evaluate and see if our data makes statistical sense)
    - Understanding variable r/ships
    - Reduce irrelevant features

#### The core idea of Hypothesis testing  
**Null Hypothesis(H<sub>0</sub>)**: This is the default assumption that there's no effect, no r/ship, no difference between group.    
_Example_  Gender and Loan approval (Nul hypothesis will be gender and loan approval are independent meaning the feature doesn't affect the target and the observed difference is by random chance)  
**Alternative Hypothesis(H<sub>1</sub>)**: This is what we are trying to prove.  
_Example_ Gender and Loan Approval are associated(meaning that theres predictive information from the Gender feature)  

__Definition of terms__  
- Confidence interval - It is a measure of how consistent the data are with the null hypothesis.  
- alpha/level of significance- Alpha is the cutoff for "statistical significance". mostly used alpha is 0.05/5%, 0.1/10%, 0.01/1%
- p-value - Is the probability of observing results atleast as extreme as the current data assuming the null hypothesis is true.  

|Condition|Decision|
|------|------|
|p < alpha|Reject H0|
|p>= alpha|Fail to reject H0|

- Type I Error(False Positive) - When we reject a null hypothesis that is true
- Type II Error(False Negative) - When we fail to reject the null hypothesis  

_How do I choose best hypothesis test_  
|Feature Type|Target Variable|Recommended test|
|------------|---------------|----------------|
|Numerical|Numerical|Correlation/Regression|
|Numerical |Categorical|t-test/ANOVA/MannuWhitney/Kruskal|
|Categorical|Categorical|Chi-Square Test|

### Numeric Features(Numeric vs Categorical target variable)  
When the feature is numerical and the target is categorical the goal is usually to determine:  
"Do the numerical values differ significantly across targets classes?"  
Null Hypothesis H0: There's no significant difference across the target classes  
Alternative Hypothesis(H1): There's significant difference across the target classes  
|Numeric Feature|Categorical target|
|---------------|------------------|
|Income|Loan Status/Approval|
|Age    |Disease Status|
|Transaction Amount|Fraud Type|
|Exam Score|Pass/Fail|    

The core statistical question is we want to determine whether the numerical feature behaves differently across categories.  
Example:  
_Does the the applicants by gender have different average incomes?_   
The choice of test to be used depends mainly on:  
- Number of target categories
    - Binary(2 classes)ttest/ztest
    - Multiclass(more than 2 classes)  ANOVA
- Sample size   
    - Use ttest if n < 30 and ztest if n>=30  
- Whether population variance is known(Variance is a statistical measurement that tells you how spread out your data is from the average/mean)
    - if the population variance is known use ztest  
- Distribution assumptions.  
    - For normal distribution you can use ttest/ANOVA/Ztest
    - If your data is not normally distributed(Binary use MannuWhitney : Multiclass use Kruskal)  

|Situation                  |Test            |
|---------------------------|----------------|
|Numerical vs Binary Target|t-test or z-test|
|Numerical vs Multiclass Target|ANOVA|
|Non-normal distributed data|Mann-Whitney(Binary)/Kruskal-Wallis(Multiclass)|  

1. __Binary Distribution Case__  
Supposed our target:   

|Loan Status|
|-----------|
|Approved|
|Rejected|  

and we have a feature called income, we compare:  
_Mean income of approved applicants vs mean income of rejected applicants_  
__Hypothesis__  
Null Hypothesis: mean1(Approved) = mean2(Rejected)    
Alternative Hypothesis: mean1(Approved) <> mean2(Rejected)  

__t-test__  
The t-test deermines whether two sample means differ significantly.  
It's the most common test in ML and data Science for:  
Numerical Features vs Binary Categorical Target.  
We use t-test because in real-world datasets, population variance is usually unknown.  
_**Types of t-tet**_  
|Type   |                                       Usage|  
|-------|--------------------------------------------|
|One-sample t-test|Compare sample mean to fixed value|
|Independent 2 sample t-test|Compares two independent groups(A/B testing)|
|Paired t-test|Compare same subjects before and after|  

For ML classification, Independent two-sample t-test is the most common.  
<br/> ttest formula<br/>  
$$t = \frac{\bar{x}_1 - \bar{x}_2}{SE}$$
 

Where:    
x1 bar = Mean1(approved)   
x2 bar = Mean2(rejected)  

___interpretation___   
_Large t-value_  
Means:  
group means differ significantly thus reject null hypothesis.  

_Small t-value_  
Means:  
group means are similar thus fail to reject null hypothesis.  

__z-test__  
A z-test is similar a t-test but it assumes:  
- Population variance is known or sample size is very large(n>=30)  

2. __Multiclass Distribution Case__ 

Suppose our target:  

|Target Group|
|------------|
|functional|
|non-functional|
|functional needs repair|  

we are comparing more than 2 group means thus we'll sue ANOVA(Analysis Of Variance):  

__Hypothesis__  
Null Hypothesis(H0): mean1=mean2=mean3  
Alternative Hypothesis: Atleast one of the means differs.  


__Categorical vs Categorical target__  
Chi-Square test of independence.   
_Why Chi-Square?_  It's designed to determine whether two categorical vriables are associated.  
It compares:  
a. What we actually observe in the data   
vs  
what we would expect if there was no r/ship  

_Intuition behind chi-square_   
Suppose we have student loan approval as below:   
|Education Level|Approved|Rejected|
|---------------|--------|--------|
|High School|20|80|
|Degree |70|30|  

We immediately notice:  
a Degree holders are approved more often
b HighSchool applicants are rejected more often  

But :  
Is this a relationship or just a random chance/coincidence?

```python

# Use the permit column 0.05, 0.01, 0.10 
alpha = 0.05  # significance level

cat_columns = ["permit","waterpoint_type_group", "source_class", "quantity", "payment_type", "management_group", "extraction_type_class", "region",	"basin"]

for column in cat_columns:
    null = f"The {column} and status group(target)  are independent"
    alternative = f"The {column} and status group are not independent"  

    contingency_table = pd.crosstab(df[column], df['status_group']) 

    chi2, p, dof, expected = chi2_contingency(contingency_table)  

    print(f"Chi sqaure test for '{column}'")
    print(f"P-Value for {column} is {p}")  
    if p < alpha:  
        print("Reject the null hypothesis", null)
        print("="*70)
    else:
        print("Fail to reject null hypothesis that ", null)
```

##### DISCLAIMER!! ANOVA test: For this project our columns are not normally distributed so ANOVA in this case is for demonstration purposes   

```python
alpha = 0.05  
null1 = "The means of numerical features are equal/same across all status_groups(mean1=mean2=mean3)"  
alternative1 = "Atleast one status_group has a different mean"  

num_features = ["amount_tsh",	"gps_height",	"population", "month_recorded",	"age"]  

for col in num_features:  
    null1 = f"The means of {col} feature is equal/same across all status_groups(mean1=mean2=mean3)"  
    alternative1 = "Atleast one status_group has a different mean" 
    groups = [group[col].values for name, group in df.groupby('status_group')]  
    stat, p = f_oneway(*groups)
    print("="* 70)  
    print(f"ANOVA test for '{col}'")  
    print(f"P-Value for {col} is {p}") 
    if p < alpha:
        print(f"Reject null hypothesis {null1}")
    else:
        print(f"Fail to reject null hypothesis{null1}")
```

#### What if the data is not normally distributed?  
- Many real world ML datasets violate the normality assumption  
- When:  
    - numerical feature is not nurmally distributed, 
    - or contains strong outliers, 
    - or samples are small, 
in such cases the standard parametric tests e.g t-test, z-test, ANOVA may become unreliable. 

__In such Cases we use:  
#### non parametric tests  
These tests do not assume normality and are more robust to skewed distributions and outliers.  

|Type|Assumes Normality?|Uses Means?| 
|----|-------------------|----------|
|Parametric|Yes|Yes|
|Non-parametric|No|Useally use ranks|  

___If Data is not normal___  
The replacement depends on:

- Binary > Mann-Whitney U test

- Multiclass > KruskalWallis test 


_When to use non-parametric tests_  
1. Numerical feature is not normally distributed.    
2. Contains strong outliers. 
3. Sample sizes are small.  

Non-parametric tests do not assume normality and they're more robust to skewed distributions and outliers.  
 

```python
# Hypothesis tests for numerical features vs categorical target variable using Kruskal-Wallis test(Multi-class classification)
alpha = 0.05 

numerical_features = ["amount_tsh",	"gps_height",	"population", "month_recorded",	"age"]


for feature in numerical_features:  
    h0 = f"The distributions of numerical feature {feature} are the same across all status_groups"  
    h1 = f"Atleast one status_group has a different distribution for {feature}"
    groups = [group[feature].values for name, group in df.groupby('status_group')]
    stat, p = kruskal(*groups)
    print("="*100)
    print(f"P-value for {feature} is {p}")
    

    if p < alpha:
        print(f"We reject the null hypothesis: {h0}")
    elif p>=alpha:
        print(f"We fail to reject the null hypothesis: {h0}")
```
